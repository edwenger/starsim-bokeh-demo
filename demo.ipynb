```python
import starsim as ss
```

```python
people = ss.People(n_agents=1)

sim = ss.Sim(people=people, diseases=[], networks=[], 
             demographics=[ss.Pregnancy(fertility_rate=500)],
             n_years=5)
sim.initialize()
```

```python
sim.pars.pregnancy
```

```python
sim.people.age[0] = 15
sim.people.female[0] = True
sim.people.person(0)
```

```python
sim.run()
sim.plot()
```
